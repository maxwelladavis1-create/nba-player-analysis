```python
# Question 6 (UPDATED)
# Restate research questions + determine needed columns (including position big or small + free throws)

# Research Questions:
# 1) How does FG% relate to scoring (PTS / PPG)?
# 2) Can lower FG% players still score more because they take more shots (FGA)?
# Teacher feedback updates:
# - Filter out players with very few minutes/zeros using thresholds (handled in Q7)
# - Position matters: simplify POS into BIG vs SMALL (handled in Q7/Q8)
# - Add free throws since they contribute to points (FTM/FTA/FT%)

needed_columns = ["PName","Team","POS","GP","Min","PTS","FGM","FGA","FG%","FTM","FTA","FT%"]
df[needed_columns].head()

```

```python
# Question 7 (UPDATED)
# Create a focused project dataset + remove low-minute/low-scoring players
# Also create SMALL vs BIG (0/1) position variable for easier modeling

import numpy as np
import pandas as pd

project_df = df[needed_columns].copy()

# Ensure numeric columns are numeric (avoid hidden strings causing issues)
num_cols = ["GP","Min","PTS","FGM","FGA","FG%","FTM","FTA","FT%"]
for c in num_cols:
    project_df[c] = pd.to_numeric(project_df[c], errors="coerce")

# Drop rows missing the core values needed for your analysis
project_df = project_df.dropna(subset=["POS","GP","Min","PTS","FGA","FG%","FTM","FTA","FT%"]).reset_index(drop=True)

#Threshold filters (teacher feedback)
# Eyeballed what a reasonable minimum minute threshold would be as a baskwtball player myself and considering 82 games
MIN_MINUTES = 400
MIN_POINTS  = 100

project_df = project_df[(project_df["Min"] >= MIN_MINUTES) & (project_df["PTS"] >= MIN_POINTS)].reset_index(drop=True)

# Create SMALL vs BIG
# Basketball note: guards = backcourt (small), forwards/centers = frontcourt (big)
# SMALL = 0, BIG = 1
pos = project_df["POS"].astype(str).str.upper().str.strip()

# BIG if contains C OR contains F (including PF/SF, etc.)
# SMALL if primarily guard-type
project_df["BIG"] = np.where(pos.str.contains("C") | pos.str.contains("F"), 1, 0).astype(int)

# Quick check
print("Rows after filtering:", project_df.shape[0])
project_df[["PName","Team","POS","BIG","Min","PTS","FGA","FG%","FTM","FTA"]].head(10)


```

```python

# Proj. assignment 3 - First data manipulation!! Processing Strings.
# Merging was not needed due to only having one file.

# explanation:

# Clean and standardize text columns (name, team, position)
# Removes spacing/symbol issues and makes categories consistent
# Splits player name into first and last name for easier analysis
# Clean/standardize key text columns


project_df["PName"] = (
    project_df["PName"]
    .astype(str)
    .str.replace("*", "", regex=False)   # common cleaning step if names include *
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

project_df["Team"] = project_df["Team"].astype(str).str.upper().str.strip()
project_df["POS"]  = project_df["POS"].astype(str).str.upper().str.strip()

# Example "parse strings" step (split player name)
# Creates two new columns: FirstName and LastName
name_parts = project_df["PName"].str.split(" ", n=1, expand=True)
project_df["FirstName"] = name_parts[0]
project_df["LastName"]  = name_parts[1].fillna("")  # handles 1-word names safely

project_df[["PName","FirstName","LastName","Team","POS"]].head(10)
```

```python
# Quick check to confirm the exact column names in project_df (error came up)
print(project_df.columns.tolist())
```

```python

# Proj assignment 3 - More data cleaning/meanipulation - Reshaping & Pivoting

# Explanation:

# Reorganizes data to compare BIG vs SMALL players by team
# Pivot shows average PPG for each group
# Melt converts it back to long format for analysis


# Find the correct "points per game" column name
possible_ppg_cols = ["PPG", "PTS", "PTS/G", "Points", "PointsPerGame"]

ppg_col = None
for c in possible_ppg_cols:
    if c in project_df.columns:
        ppg_col = c
        break

if ppg_col is None:
    raise KeyError(
        "Could not find a points-per-game column. "
        "Check project_df.columns.tolist() and update possible_ppg_cols."
    )

print("Using points column:", ppg_col)

# Pivot: average points per game by Team and BIG
ppg_pivot = project_df.pivot_table(
    index="Team",
    columns="BIG",
    values=ppg_col,
    aggfunc="mean"
)

ppg_pivot = ppg_pivot.rename(columns={0: "Avg_PPG_SMALL", 1: "Avg_PPG_BIG"}).reset_index()
display(ppg_pivot.head(10))

# Melt back to long format
ppg_long = pd.melt(
    ppg_pivot,
    id_vars="Team",
    value_vars=["Avg_PPG_SMALL", "Avg_PPG_BIG"],
    var_name="PlayerGroup",
    value_name="Avg_PPG"
)

display(ppg_long.head(10))
```

```python
# Question 8 (UPDATED)
# Determine analyses and compute features for correlations and modeling
# Includes grouping by BIG vs SMALL and adds free throws to the model

import statsmodels.api as sm

# Create per-game metrics - better comparison stat
project_df["PPG"] = project_df["PTS"] / project_df["GP"]
project_df["FGA_PG"] = project_df["FGA"] / project_df["GP"]
project_df["FTA_PG"] = project_df["FTA"] / project_df["GP"]
project_df["FTM_PG"] = project_df["FTM"] / project_df["GP"]
project_df["MIN_PG"] = project_df["Min"] / project_df["GP"]

# 1) Overall correlations
project_df[["PPG","FG%","FGA_PG","FTA_PG","MIN_PG","BIG"]].corr(numeric_only=True)


```

```python
# Assignment 6 Advanced Analysis
# Exploring scoring efficiency and shot volume.

# 1. Create a scoring efficiency metric (points per shot attempt).
# 2. Compare efficiency between BIG and SMALL players.
# 3. Compare efficiency between high-volume and low-volume shooters.
```

```python
# Analysis 1: Create Efficiency Metric

# Points per Shot Attempt (efficiency metric)
# Uses total points divided by total field goal attempts

project_df["PTS_per_Shot"] = project_df["PTS"] / project_df["FGA"]

# Check results
project_df[["PName","PTS","FGA","PTS_per_Shot"]].head()
```

```python
# Analysis 2: Efficiency Comparison (BIG vs SMALL)

# Group by BIG vs SMALL
efficiency_by_size = project_df.groupby("BIG")["PTS_per_Shot"].mean()

print("Average Points per Shot Attempt (Efficiency):")
print(efficiency_by_size)

# Visualization
plt.bar(efficiency_by_size.index.astype(str), efficiency_by_size.values)

plt.title("Scoring Efficiency: BIG vs SMALL Players")
plt.xlabel("Player Group (0 = SMALL, 1 = BIG)")
plt.ylabel("Average Points per Shot Attempt")

plt.show()
```

```python
# Analysis 3: Create High vs Low Volume Shooters


# Determine the median shot attempts per game
median_fga = project_df["FGA_PG"].median()

# Create new volume group variable
project_df["Volume_Group"] = np.where(
    project_df["FGA_PG"] >= median_fga,
    "High Volume",
    "Low Volume"
)

# Check counts
project_df["Volume_Group"].value_counts()
```

```python
# Analysis 4: Efficiency Comparison (Shot Volume)

efficiency_by_volume = project_df.groupby("Volume_Group")["PTS_per_Shot"].mean()

print("Average Points per Shot Attempt by Volume Group:")
print(efficiency_by_volume)

# Visualization
plt.bar(efficiency_by_volume.index, efficiency_by_volume.values)

plt.title("Scoring Efficiency: High vs Low Volume Shooters")
plt.xlabel("Shot Volume Group")
plt.ylabel("Average Points per Shot Attempt")

plt.show()
```

```python
# Analysis 5: Efficiency by Position AND Shot Volume

combined_efficiency = project_df.groupby(
    ["BIG", "Volume_Group"]
)["PTS_per_Shot"].mean().unstack()

print("Efficiency by Player Type and Shot Volume:")
print(combined_efficiency)

# Plot
combined_efficiency.plot(kind="bar")

plt.title("Efficiency by Player Type and Shot Volume")
plt.xlabel("Player Group (0 = SMALL, 1 = BIG)")
plt.ylabel("Average Points per Shot Attempt")

plt.show()
```

```python
# 3) Regression model: Predict PPG using FG% + shot volume + free throws + minutes + BIG (0/1)

X = project_df[["FG%","FGA_PG","FTA_PG","MIN_PG","BIG"]].copy()

# Ensure purely numeric for statsmodels
X = X.apply(pd.to_numeric, errors="coerce")

X = sm.add_constant(X)
y = project_df["PPG"]

model = sm.OLS(y, X, missing="drop").fit()
model.summary()
```
